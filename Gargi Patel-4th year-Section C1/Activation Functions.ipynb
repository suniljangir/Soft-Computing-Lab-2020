```python
import matplotlib.pyplot as plt 
import numpy as np 
import math 

#SIGMOID
def sigmoid(x):
    z = 1/(1 + np.exp(-x)) 
    return z
x = np.linspace(-10, 10, 100)   
plt.plot(x,sigmoid(x)) 
plt.xlabel("x-axis") 
plt.ylabel("Sigmoid(X)") 
```

```python
#Relu
def relu(x): 
    return np.maximum(0,x)
x = np.linspace(-10, 10, 100)   
plt.plot(x,relu(x)) 
plt.xlabel("x-axis") 
plt.ylabel("relu(x)")   
plt.show()
```

```python
#Tanh

def tanh(x):
    t=(np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
    return t
z=np.arange(-4,4,0.01)
plt.plot(z,tanh(z), color="#307EC7", linewidth=3, label="tanh")
plt.xlabel("x-axis")
plt.ylabel("tanh(x)")
plt.show()

```

```python
#Softmax

def softmax(X):
    expo = np.exp(X)
    expo_sum = np.sum(np.exp(X))
    return expo/expo_sum
x = np.linspace(-10, 10, 100)   
plt.plot(x,softmax(x)) 
plt.xlabel("x-axis") 
plt.ylabel("softmax(x)")   
plt.show()
```
