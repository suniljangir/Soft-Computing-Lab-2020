AIM: Create plot for given equations.

```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

1. y=e^ax for different values of a

```
import random
x = np.linspace(-10, 10, 100) 
a=np.random.random()
y = np.exp(a*x)
fig = plt.figure(figsize = (10,5)) 
plt.grid(alpha =.7, linestyle ='--') 
plt.xlabel('x-axis') 
plt.ylabel('y-axis') 
plt.plot(x,y,'m',linewidth=2)
plt.show()
```

2) y=7x^2 + 3x + 10 for 2<=x<=5

```
x = np.linspace(2, 5, 100) 
y = 7*x**2 + 3*x + 10  
fig = plt.figure(figsize = (10, 5)) 
plt.plot(x, y,'r',alpha=0.8,linestyle ='--') 
plt.grid(alpha =.7, linestyle ='--') 
plt.xlabel('x-axis') 
plt.ylabel('y-axis') 
plt.show() 
```

3) y=1/1+e^-x


```
x = np.linspace(-5, 5, 100) 
y = 1/(1+np.exp(-x))
fig = plt.figure(figsize = (10,5)) 
plt.plot(x,y,'y',linestyle='-.')
plt.grid(alpha =.7, linestyle ='--') 
plt.xlabel('x axis') 
plt.ylabel('y axis') 
plt.show()
```

4. y=1-e^-ax/1+e^ax for different values of a

```
import random
x = np.linspace(-30, 30, 100) 
a=np.random.random()
y = (1-np.exp(-a*x))/(1+np.exp(-a*x))
fig = plt.figure(figsize = (10,5)) 
plt.grid(alpha =.7, linestyle ='--') 
plt.xlabel('x axis') 
plt.ylabel('y axis') 
plt.plot(x,y,'g')
plt.show()
```

5. y=tan hx

```
x = np.linspace(-6, 6, 50)   
y = np.tanh(x) 
fig = plt.figure(figsize = (10,5)) 
plt.plot(x,y,'c',linestyle=':',linewidth=2)
plt.grid(alpha =.7, linestyle ='--') 
plt.xlabel('x axis') 
plt.ylabel('y axis') 
plt.show()
```
