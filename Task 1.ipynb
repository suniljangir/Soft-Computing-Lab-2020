# 2.) Operators in Python

```python
## 1. Arithmetic Operators:

x=2
y=1
a=x+y
b=x-y
c=x*y
d=x/y
e=x%y
f=x**y
print(a,",",b,",",c,",",d,",",e,",",f)
```

```python
## 2. Relational Operators

a = 13
b = 33   
print(a > b) 
print(a < b) 
print(a == b) 
print(a != b) 
print(a >= b) 
print(a <= b) 
```

```python
## 3. Logical Operators

a = True
b = False 
print(a and b)  
print(a or b)   
print(not a) 
```

```python
## 4. Bitwise Operators

a = 10
b = 4   
print(a & b) 
print(a | b)   
print(~a)  
print(a ^ b)   
print(a >> 2)   
print(a << 2)
```

```python
## 5. Assignment Operators

a = 21
b = 10
c = 0
c = a + b
print("Value of c is ", c)
c += a
print("Value of c is ", c) 
c *= a
print("Value of c is ", c) 
c /= a 
print("Value of c is ", c) 
c  = 2
c %= a
print("Value of c is ", c)
c **= a
print("Value of c is ", c)
c //= a
print("Value of c is ", c)
```
