### AIM: Write a program to implement linear separability for different functions.

1.AND

```python
import numpy as np
import matplotlib 
from matplotlib import pyplot as plt
x = np.array([0,1,0])
y = np.array([0,0,1])
plt.scatter(x,y,c='red')
plt.scatter(1,1,c="blue")
plt.xlabel('Input 1')
plt.ylabel('Input 2')
w=-1
b=1.5
x = np.linspace(0,1.5)
plt.plot(x,w*x+b,c='black')
plt.show()
```

2.OR

```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
x = np.array([0,1])
y = np.array([0,1])
plt.scatter(x,y,c='red')
x = np.array([1,0])
y = np.array([0,1])
plt.scatter(x,y,c="blue")
plt.xlabel('Input 1')
plt.ylabel('Input 2')
w=-1
b=1.5
x = np.linspace(0,1.5)
plt.plot(x,w*x+b,c='black')
plt.show()
```
