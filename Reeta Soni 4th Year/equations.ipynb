y={1 x<10 0 else}

```python
import matplotlib.pyplot as plt
```

```python
x=[num for num in range(0,20)]
y=[]
for i in x:
    if i>=10:
        y.append(1)
    else:
        y.append(0)
```

```python
plt.plot(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("y=1 if x>=10 else y=0")
plt.show()
```

y=eax, for different values of a.

```python
import numpy as np
```

```python
num=[num for num in range(0,10)]
a=2 #constant value
x=np.array(num)*2
y=np.exp(x)
```

```python
plt.plot(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("y=e**ax")
plt.show()
```

y=7x2+3x+10 for 2<=x<=5

```python
x=[x for x in range(2,6)]
y=[]
```

```python
for i in x:
    y.append(7*pow(i,2)+3*i+10)
```

```python
plt.plot(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('y=7x^2+3x+10')
plt.show()
```

y=1/1+e^-x

```python
x=[x for x in range(0,20)]
p=np.array(x)
p=1+np.exp(-p)
y=1/p
```

```python
plt.plot(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

y=1-e^-ax/1+e^-ax, for different values of a.

```python

num=[num for num in range(0,21)]
num=np.array(num)
a=2
p=1-np.exp(-num*a)
q=1+np.exp(-num*a)
y=p/q
```

```python
plt.plot(num,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```
